```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.mygreatlearning.com/blog/top-startups-in-india/'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')
```

```python
print(table)
```

```python
t_header = soup.find_all('th')
```

```python
print(t_header)
```

```python
table_titles = [title.text.strip() for title in t_header]

print(table_titles)
```

```python
import pandas as pd
```

```python
data_frame = pd.DataFrame(columns=table_titles)

data_frame
```

```python
column_data = soup.find_all('tr')
```

```python
for rows in column_data[1:]:
    row_data = rows.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    #print(individual_row_data)
    length = len(data_frame)
    data_frame.loc[length] = individual_row_data
```

```python
data_frame
```
